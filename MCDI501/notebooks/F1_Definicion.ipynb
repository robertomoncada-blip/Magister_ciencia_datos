En la siguiente celda se ejecuta el código que lee el archivo csv y lo carga a un objeto de tipo DataFrame del paquete Pandas. Además, se muestra la cantidad de registros del Dataset y la cantidad de variables incluidas, junto al nombre y tipo de dato de las caracteristicas (variable)

```python
import sys
import os
import importlib
import pandas as pd

# 1. Configurar la ruta primero
ruta = os.path.abspath("../src")
if ruta not in sys.path:
    sys.path.append(ruta)

# 2. Importar el módulo completo
import data_loading

# 3. Recargar (ahora sí Python sabe qué es data_loading)
importlib.reload(data_loading)

# 4. Importar la función específica
from data_loading import cargar_datos

# 5. Ejecutar
ds = cargar_datos('../data/raw/UCI_Credit_Card.csv')

# 6. Identificar tipos y valores nulos
print("="*55)
print("Identificar tipos y valores nulos")
print("="*55)
print(ds.info())

# 7. Visualizar primeras filas
print("="*55)
print("Visualizar primeras filas")
print("="*55)
print(ds.head())

# 8. Verificar coherencia de tipos
print("="*55)
print("Verificar coherencia de tipos")
print("="*55)
print(ds.dtypes) 

# 9. ELIMINAR columna ID (no aporta valor predictivo)
ds = ds.drop(columns=['ID'])

# 10. VARIABLES CATEGÓRICAS (int64 → category)
categoricas = ['SEX', 'EDUCATION', 'MARRIAGE',
               'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
ds[categoricas] = ds[categoricas].astype('category')

# 11. VARIABLE TARGET (int64 → uint8)
ds['default.payment.next.month'] = ds['default.payment.next.month'].astype('uint8')

# 12. VARIABLES MONETARIAS (float64 → float32)
float_cols = ['LIMIT_BAL',
              'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3',
              'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
              'PAY_AMT1',  'PAY_AMT2',  'PAY_AMT3',
              'PAY_AMT4',  'PAY_AMT5',  'PAY_AMT6']
ds[float_cols] = ds[float_cols].astype('float32')

# 13. EDAD (int64 → uint8, no hay edades negativas ni > 255)
ds['AGE'] = ds['AGE'].astype('uint8')

# 14. VERIFICAR resultado
print(ds.dtypes)
print(f"\nMemoria usada: {ds.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

#15 Analizar distribuciones
print("="*55)
print("Análisis de distribuciones")
print("="*55)
print(ds.describe(include='all'))     

#16 Detectar valores faltantes
print("="*55)
print("Detectar valores faltantes")
print("="*55)
print(ds.isnull().sum())

#17 Analizar valores únicos por columna
# ── 17.1 RESUMEN GENERAL: cantidad de valores únicos por columna ──────────────
print("="*55)
print("VALORES ÚNICOS POR COLUMNA")
print("="*55)
print(ds.nunique().to_string())

# ── 17.2 VALUE_COUNTS POR COLUMNA CATEGÓRICA ─────────────────────────────────
print("\n" + "="*55)
print("DISTRIBUCIÓN DE CATEGORÍAS")
print("="*55)

categoricas = ['SEX', 'EDUCATION', 'MARRIAGE',
               'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6',
               'default.payment.next.month']

for col in categoricas:
    print(f"\n📊 {col}")
    print("-"*35)
    vc = ds[col].value_counts(dropna=False).sort_index()
    pct = ds[col].value_counts(normalize=True, dropna=False).sort_index() * 100
    resumen = pd.DataFrame({'Frecuencia': vc, 'Porcentaje (%)': pct.round(2)})
    print(resumen.to_string())

# ── 17.3 RESUMEN RÁPIDO VARIABLES NUMÉRICAS CONTINUAS ────────────────────────
print("\n" + "="*55)
print("VALORES ÚNICOS - VARIABLES CONTINUAS")
print("="*55)

continuas = ['LIMIT_BAL', 'AGE',
             'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3',
             'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
             'PAY_AMT1',  'PAY_AMT2',  'PAY_AMT3',
             'PAY_AMT4',  'PAY_AMT5',  'PAY_AMT6']

resumen_cont = pd.DataFrame({
    'Únicos'  : ds[continuas].nunique(),
    'Mínimo'  : ds[continuas].min(),
    'Máximo'  : ds[continuas].max(),
    'Media'   : ds[continuas].mean().round(2)
})
print(resumen_cont.to_string())
```
